Define functions

```python
def get_intensity (input_file):
    # Reads file and extracts intensity
    file_pointer = open(input_file, 'r').readlines()
    i=0
    for line in file_pointer:
        if '# External' not in line:
            i+=1
        else:
            break
    skip_lines=len(file_pointer)-i
    output=np.genfromtxt(input_file,skip_footer=skip_lines)
    intensity=np.array([elem[1] for elem in output])
    return intensity

def get_region (input_file):
    '''
    Looks in comments and returns run number
    '''
    run_info=get_run_info(input_file)
    if run_info.find('_'):
        start=run_info.find(':')
        end=run_info.find('_')
        region=run_info[start+1:end]
    else: run_info=None
    return region

def get_run_nr (input_file):
    '''
    Looks in file and returns excitation energy
    '''
    run_info=get_run_info(input_file)
    if run_info.find('_'):
        l=run_info.find('_')
        run_nr=run_info[l+1:]
    else:
        run_nr=None
    return run_nr

def get_scan_mode (input_file):
    '''
    Looks in comments and returns if the scan is XAS or XPS
    '''
    with open(input_file, 'r') as file:  
        file_contents=file.readlines()
        scan_mode=file_contents[21].strip('# Scan Mode:         ').rstrip('\n')
    if scan_mode:
        if scan_mode=='FixedAnalyzerTransmission':
            return 'XPS'
        else:
            return 'XAS'
    else: return None

def get_run_info (input_file):
    '''
    Looks in comments and returns the region info from line 15
    '''
    with open(input_file, 'r') as file:  
        file_contents=file.readlines()
        run_info=file_contents[15].strip('# Region:            ').rstrip('\n')
    return run_info

def rename_files (input_file,base_path):
    run_nr=get_run_nr(base_path+input_file)
    region=get_region(base_path+input_file)
    scan_mode=get_scan_mode(base_path+input_file)
    new_file_name = base_path+'work/'+run_nr+'_'+scan_mode+'_'+region+'.txt'
    # print(input_file)
    with open(base_path+input_file) as file:
        if not os.path.exists(os.path.dirname(new_file_name)):
            os.makedirs(os.path.dirname(new_file_name))
        if not os.path.isfile(base_path+new_file_name):
            new_file = open(new_file_name,'w+')
            new_file.write(file.read())
            new_file.close()
        elif os.path.isfile(base_path+new_file_name):
            new_file_name_1=base_path+'work/'+run_nr+'_'+scan_mode+'_'+region+'_01.txt'
            new_file = open(new_file_name_1,'w+')
            new_file.write(file.read())
            new_file.close()
    pass
    
def discover_files(path):
    '''
    Looks in the given directory and returns the filenames
    '''
    for (dirpath, dirnames, filenames) in walk(path):
        break
    return dirnames,filenames
```
